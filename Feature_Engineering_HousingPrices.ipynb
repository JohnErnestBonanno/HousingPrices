```python
#Feature Engineering

#Going to try and create a few features that make sense to me.  Should condense data and help with EDA
```

```python
#Count total number of Full and Half Baths
dfMaster['TotFullBath'] = (dfMaster['BsmtFullBath'] + dfMaster['FullBath'])
dfMaster['TotHalfBath'] = (dfMaster['BsmtHalfBath'] + dfMaster['HalfBath'])
```

```python
#Age of House (data set is from 2011)
dfMaster['HouseAge'] = (2011 -dfMaster['YearBuilt'])

#Time since last Remodel
dfMaster['YrSinceRemod'] = (2011 - dfMaster['YearRemodAdd'])
```

```python
#Total Square Feet
dfMaster['TotSF'] = (dfMaster['1stFlrSF'] + dfMaster['2ndFlrSF'])
```

```python
#Think I can use the same function over and over again.  
```

```python
def HasFunction(x):
    if x > 0:
        return(1)
    else:
        return(0)
```

```python
dfMaster['HasBasement'] = dfMaster['BsmtFinSF1'].apply(HasFunction)
dfMaster['Has2Floor'] = dfMaster['2ndFlrSF'].apply(HasFunction)
dfMaster['HasGarage'] = dfMaster['GarageArea'].apply(HasFunction)
dfMaster['HasPool'] = dfMaster['PoolArea'].apply(HasFunction)
```
